```python
import matplotlib.pyplot as plt

# data
x = [4, 8, 16, 32, 64]
y = [.78, 2.5, 9.5, 32.3, 65.1]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('# nodes')
ax.set_ylabel('First Value')
ax.set_title('Beacon Generation Plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [16, 32, 64]
y1 = [5, 20, 45]
y2 = [1, 3, 8]

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the first dataset using the left y-axis (ax1)
ax1.plot(x, y1, label='Y1', color='tab:blue')
ax1.set_xlabel('# nodes')
# ax1.set_ylabel('Y1', color='tab:blue')

# Create a second y-axis on the right side (ax2)
ax2 = ax1.twinx()

# Plot the second dataset using the right y-axis (ax2)
ax2.plot(x, y2, label='Y2', color='tab:orange')
# ax2.set_ylabel('Y2', color='tab:orange')

# Add a legend
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Set a title
plt.title('Setup Phase')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [16, 32, 64]
y1 = [9, 32, 65]
y2 = [9, 34, 88]

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot the first dataset using the left y-axis (ax1)
ax1.plot(x, y1, label='Y1', color='tab:blue')
ax1.set_xlabel('# nodes')
ax1.set_ylabel('Y1', color='tab:blue')

# Plot the second dataset using the same y-axis (ax1)
ax1.plot(x, y2, label='Y2', color='tab:orange')
ax1.set_ylabel('Y2', color='tab:orange', labelpad=20)  # Adjust labelpad to move the label down

# Add legends with an empty label to create spacing
ax1.legend(handles=[plt.Line2D([], [], color='tab:blue'), plt.Line2D([], [], color='tab:orange')],
            labels=['GRandLine', 'BRandPiper'], loc='upper left')

# Set a title
plt.title('Beacon Generation Efficiency Plot')

# Show the plot
plt.show()
```

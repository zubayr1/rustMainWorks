```python
import matplotlib.pyplot as plt

# data
x = [16, 32, 64]
y = [0.8, 3, 8]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('# nodes')
ax.set_ylabel('first value')
ax.set_title('BRandPiper Plot')

# Show the plot
plt.show()
```

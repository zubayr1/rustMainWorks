```python
import matplotlib.pyplot as plt

# data
x = [4, 8, 16]
y = [4, 7, 16]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add labels and a title
ax.set_xlabel('# nodes')
ax.set_ylabel('first value')
ax.set_title('GRandLine Plot')

# Show the plot
plt.show()
```
